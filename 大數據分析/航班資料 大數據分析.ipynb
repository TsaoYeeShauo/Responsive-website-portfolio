```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('./flights.csv')
flights.head(10)
```

```python
#Grouping and aggregating with multiple columns and functions
#For each origin to destination flight, find the total number of flights, the number and percentage of canceled flights and the variance of the airline. 
group_cols=['ORG_AIR','DEST_AIR']
agg_dict={'CANCELLED':['sum','mean','size'],'AIR_TIME':['mean','std']}
flights.groupby(group_cols).agg(agg_dict).head()
```

```python
#Removing the MultiIndex after grouping
#Read in the flights dataset; write a statement to find the total and average miles flown; and the maximum 
#and minimum arrival delay for each airline for each weekday
airline_info = flights.groupby(['AIRLINE','WEEKDAY']).agg({'DIST':['sum','mean'],'ARR_DELAY':['max','min']}).astype(int)
airline_info.head()   
```

```python
#Let's see an example of this by finding the average distance traveled per flight from each airline
flights.groupby(['AIRLINE'], as_index=False)['DIST'].agg('mean').round(0)
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].agg('mean').head()
```
