```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv('./employee.csv')
depts = ['Houston Police Department-HPD', 'Houston Fire Department (HFD)']
select_columns = ['UNIQUE_ID','DEPARTMENT','GENDER','BASE_SALARY']
```

```python
employee.DEPARTMENT.value_counts().head()
```

```python
#define the selection criteria
depts = ['Houston Police Department-HPD', 'Houston Fire Department (HFD)']
criteria_dept = employee.DEPARTMENT.isin(depts)
criteria_gender = employee.GENDER == 'Female'
criteria_salary = (employee.BASE_SALARY>=80000) & (employee.BASE_SALARY <= 120000)
criteria_final = criteria_dept & criteria_gender & criteria_salary
```

```python
#Now let's get query result
select_columns = ['UNIQUE_ID','DEPARTMENT','GENDER','BASE_SALARY']
employee.loc[criteria_final, select_columns].head()
```

```python
#we can define criteria_salary with between() method
criteria_salary 
criteria_final
select_columns
```

```python
# Build the query string and execute the method
qs = "DEPARTMENT in @depts " \
     "and GENDER == 'Female' " \
     "and 80000 <= BASE_SALARY <= 120000"
```

```python
emp_filtered = employee.query(qs)
emp_filtered[select_columns].head() 
```

```python
#Instead of manually typing in a list of department names,
#we could have programmatically created it
#For instance, if we wanted to find all the female
#employees that were not a member of the top 10
#departments by frequency, we can run the following code
top10_depts = employee.DEPARTMENT.value_counts().index[:5].tolist()
qs = "DEPARTMENT not in @top10_depts and GENDER == 'Female' "
employee_filtered2 = employee.query(qs)
employee_filtered2[['DEPARTMENT','GENDER']].head()
```
