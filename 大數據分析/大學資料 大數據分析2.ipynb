Replicating boolean indexing with
index selection


```
#Read in the college dataset and use boolean indexing
#to select all institutions from the state of Texas (TX):
college = pd.read_csv('./college.csv')
college[college['STABBR']=='TX'].head()
```

```
#To replicate this using index selection, we need to
#move the STABBR column into the index.
college2 = college.set_index('STABBR')
#We can then use label-based selection with the .loc indexer
college2.loc['TX'].head()
```

```
#Let's compare the speed of both methods
%timeit college[college['STABBR']=='TX']
```

```
#Let's compare the speed of both methods
%timeit college2.loc['TX']
```

Example2 Let's select Texas (TX), California (CA), and New York(NY).
• With boolean selection, you can use the isin method but
with indexing, just pass a list to .loc

```
states = ['TX','CA','NY']
college[college['STABBR'].isin(states)]
college2.loc[states].head()
```

```
states = ['TX','CA','NY']
college[college['STABBR'].isin(states)]
```
