```python
#we will use only 6 columns of college dataframe
different_cols = ['Row ID', 'Postal Code', 'Quantity', 'Sales', 'Discount','Profit']
# we create a new dataframe: SuperStoreSales2
SuperStoreSales2 = SuperStoreSales.loc[:, different_cols]
# let's check out the top 5 records of SuperStoreSales2
SuperStoreSales2.head(5)
```

```python
SuperStoreSales2['Postal Code']=SuperStoreSales['Postal Code'].astype(np.int8)
```

```python
SuperStoreSales2['Quantity']=SuperStoreSales['Quantity'].astype(np.int8)
```

```python
#Please change the data type of “Sales”, “Discount”,
#and “Profit” to float 16.
SuperStoreSales2['Sales']=SuperStoreSales['Sales'].astype(np.int16)
```

```python
SuperStoreSales2['Discount']=SuperStoreSales['Discount'].astype(np.int16)
```

```python
SuperStoreSales2['Profit']=SuperStoreSales['Profit'].astype(np.int16)
```
