
# Customizing an aggregation function

```python
college.groupby('STABBR')['UGDS'].agg('mean','std').round(0).head()
```

```python
def max_deviation(s):
  std_score=(s - s.mean())/s.std()
  return std_score.abs().max()
```

```python
college.groupby('STABBR')['UGDS'].agg(max_deviation).round(0).head()
```

```python
college.groupby(['STABBR','RELAFFIL'])['UGDS','SATVRMID','SATMTMID']\
.agg([max_deviation,'mean','std']).round(1).head()
```

# Speeding up scalar selection

```python
#Pass a college name and column name to .loc in order to
#Select a scalar value 
cn = 'Texas A & M University-College Station'
college.loc[cn, 'UGDS_WHITE']
```

```python
#Use the %timeit magic command to find the difference in speed
%timeit college.loc[cn,'UGDS_WHITE']
```

```python
#Find the integer 
#Find the integer locations of the preceding selections and then time the difference between .iloc and .iat:
row_num = college.index.get_loc(cn)
col_num = college.columns.get_loc('UGDS_WHITE')
row_num,col_num
```

```python
college.iloc[row_num,col_num]
```

```python
college.iat[row_num,col_num]
```

```python
#get_ipython().run_line_magic('timeit','college.iloc[row_num,col_num]')
%timeit college.iloc[row_num,col_num]
```

```python
%timeit college.iat[row_num,col_num]
```
